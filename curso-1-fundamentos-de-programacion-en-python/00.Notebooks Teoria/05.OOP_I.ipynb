## Clases (``class``)

La programacin orientada a objetos (**Object-Oriented Programming (OOP)**) se basa en el hecho de que **se debe dividir el programa en modelos de objetos fsicos o simulados**. Se debe expresar un programa como un **conjunto de objetos que colaboran entre si mismos** para realizar tareas.

**Un objeto** es la representacin en un programa de un concepto y contiene la informacin necesara para abstraerlo:

- **Atributos**: que describen al objeto.
- **Mtodos**: operaciones que se pueden realizar al objeto.


Los objetos pueden agruparse en categoras y una clase describe (de un modo abstracto) todos los objetos de un tipo o categora determinada.



**La idea de la programacin orientada a objetos puede parecer abstracta y compleja**, pero ya hemos estado utilizando objetos sin darnos cuenta. **Casi todo en Python es un objeto**, todas las cadenas, listas, y diccionarios que hemos visto hasta ahora y que hemos usado han sido objetos.

- **Objeto**
    - El objeto es el centro de la programacin orientada a objetos. Un objeto es algo que se visualiza, se utiliza y que juega un papel o un rol en el dominio del problema del programa. La estructura interna y el comportamiento de un objeto, en consecuencia, no es prioritario durante el modelado del problema.
    

- **Clase**
    - En el mundo real existen varios objetos de un mismo tipo o clase, por lo que una clase equivale a la generalizacin de un tipo especfico de objetos. Una clase es una plantilla que define las variables y los mtodos que son comunes para todos los objetos de un cierto tipo.
    

- **Instancia**
    - Una vez definida la clase se pueden crear objetos a partir de sta, a este proceso se le conoce como crear instancias de una clase o instanciar una clase. En este momento el sistema reserva suficiente memoria para el objeto con todos sus atributos.

    - Una instancia es un elemento de una clase (un objeto). Cada uno de los objetos o instancias tiene su propia copia de las variables definidas en la clase de la cual son instanciados y comparten la misma implementacin de los mtodos. Sin embargo, cada objeto asigna valores a sus atributos y es totalmente independiente de los dems.
    

- **Mtodo**
    - Los mtodos especifican el comportamiento de la clase y sus instancias. En el momento de la declaracin hay que indicar cul es el tipo del parmetro que devolver el mtodo.
    

- **Atributos**
    - Tipos de datos asociados a un objeto (o a una clase de objetos), que hace los datos visibles desde fuera del objeto y esto se define como sus caractersticas predeterminadas, cuyo valor puede ser alterado por la ejecucin de algn mtodo.

### ``.__init__()``

El mtodo **``.__init__()``** (**initialize**) es **utilizado cada vez que se crea una instancia de una clase**.
Este mtodo **se utiliza para inicializar atributos** y solo es utilizado para eso, no tiene otros usos.

Al utilizar este mtodo cada vez que inicialicemos una instancia **tendremos que dar parmetros al objeto para que pueda crear los atributos, de esta forma cada instancia es tendr valores diferentes para cada instancia**.

Este mtodo **``.__init__()``** puede tomar tantos parmetros como queramos pero **el primer parmetro debe ser una variable llamada _self_** (por convencin) que har referencia a "l mismo". (Por eso el nombre **"self"**).

Luego de esto, utilizamos los parmetros de la clase para inicializar los atributos:

```python
self.atributo = parmetro
```
Aqu usamos la variable _**self**_ para que la clase entienda que el atributo se le va a agregar a esa instancia, no a todas las instancias.

Aunque no es obligatorio, por lo general **atributo** y **parmetro** comparten el mismo nombre. 

````python
class Empleado:
    
    puesto_trabajo = "Empleado"
    
    def __init__(self, nombre, apellido):
        self.nombre = nombre
        self.apellido = apellido
        self.email = f"{nombre.lower()}.{apellido.lower()}@python.com"        
        
# En este ejemplo, cuando inicialicemos una instancia tendremos que dar los parametros de nombre y apellido
# La instancia creada tendra los atributos de nombre, apellido y email variables (dependiendo de los valores)
# Y tendra el atributo "puesto_trabajo" constante, es decir, todas las instancias creadas tendrn el mismo valor.
````

````python
persona = Empleado(nombre="Pablo", apellido="Rodriguez")

persona
````

````python
persona.nombre
````

````python
persona.apellido
````

````python
persona.email
````

````python
persona.puesto_trabajo
````

````python
# Tambin podemos modificar los atributos aunque los hayamos inicializado con __init__

persona.nombre = "pablo"

persona.nombre
````

### Mtodos

**Los mtodos son funciones propias de cada clase**, es decir, estos mtodos solo funcionan con su clase.

**Para crear mtodos, basta con definir una funcin dentro de la clase**.

Estas funciones pueden tomar o no parmetros, sin embargo, si queremos usar el mtodo sobre la instancia, **es obligatorio que el primer parmetro de los mtodos sea self** (aunque sea el nico).

- En caso de querer utilizar atributos del mtodo **``.__init__()``** o **atributos estticos**, podemos simplemente usar **self** como parmetro del nuevo mtodo y con esto podremos usar cualquier atributo de la clase (Ya sea de **``.__init__()``** o **esttico**).

````python
class Empleado:
    
    puesto_trabajo = "Empleado"
    
    def __init__(self, nombre, apellido):
        self.nombre = nombre
        self.apellido = apellido
        self.email = f"{nombre.lower()}.{apellido.lower()}@python.com"
        
    def display_info(self):
        print(f"Nombre: {self.nombre}")
        print(f"Apellido: {self.apellido}")
        print(f"Contacto: {self.email}")        
        print(f"Puesto de trabajo: {self.puesto_trabajo}")
        
    def display_message(self):
        print("Esta es la clase Empleado.")
        
    def cambiar_puesto_trabajo(self, nuevo_puesto):
        self.puesto_trabajo = nuevo_puesto
````

````python
persona = Empleado(nombre="Pablo", apellido="Rodriguez")

persona
````

````python
# Mtodo .display_info()

persona.display_info()
````

````python
# Mtodo .display_message()

persona.display_message()
````

````python
# Mtodo .cambiar_puesto_trabajo()

persona.cambiar_puesto_trabajo(nuevo_puesto="Profesor")

persona.puesto_trabajo
````

````python
# Mtodo .display_info()

persona.display_info()
````

**Como los mtodos son, en esencia, funciones, podemos hacer que nos retornen algn valor.**

````python
# Aadimos el mtodo "empleado_info_dict()"

class Empleado:
    
    puesto_trabajo = "Empleado"
    
    def __init__(self, nombre, apellido):
        self.nombre = nombre
        self.apellido = apellido
        self.email = f"{nombre.lower()}.{apellido.lower()}@python.com"
        
    def empleado_info_dict(self):
        empleado_dict = {"nombre"         : self.nombre,
                         "apellido"       : self.apellido,
                         "email"          : self.email,
                         "puesto_trabajo" : self.puesto_trabajo}
        
        return empleado_dict
    
    def display_info(self):
        print(f"Nombre: {self.nombre}")
        print(f"Apellido: {self.apellido}")
        print(f"Contacto: {self.email}")        
        print(f"Puesto de trabajo: {self.puesto_trabajo}")
        
    def display_message(self):
        print("Esta es la clase Empleado.")
        
    def cambiar_puesto_trabajo(self, nuevo_puesto):
        self.puesto_trabajo = nuevo_puesto
````

````python
persona = Empleado(nombre="Pablo", apellido="Rodriguez")
````

````python
persona.empleado_info_dict()
````

````python
dir(Empleado)
````

**Ejemplo 2:**

````python
class Animal:
    
    def __init__(self, color, n_patas, domestico = False):
        self.color = color.title()
        self.n_patas = n_patas
        self.domestico = domestico
    
    def display_info(self):
        print("Color:", self.color)
        print("Numero de patas:", self.n_patas)
        print("Animal Domstico:", self.domestico)
        
        try:
            print("Nombre:", self.nombre)
            
        except:
            print("No tiene nombre.")
       
    def mascota(self):
        
        if self.domestico == True:
            nombre = input("Nombre de la mascota: ")
            self.nombre = nombre.title()
            
        else:
            print("Es un animal salvaje. No puede tener nombre.")
````

````python
perro = Animal(color="marron", n_patas=4, domestico=True)

perro.display_info()
````

````python
perro.mascota()
````

````python
perro.display_info()
````

````python
perro = Animal(color="marron", n_patas=4, domestico=False)

perro.display_info()
````

````python
perro.mascota()
````
